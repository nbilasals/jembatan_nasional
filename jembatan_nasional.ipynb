```python
import pandas as pd

# Group the data by province
grouped_data = data.groupby('Provinsi')

# Calculate statistics for each province
province_statistics = grouped_data[['Jml_Baik', 'Jml_RR', 'Jml_RB', 'Jml_Kritis', 'Jml_Runtuh/Putus']].sum()

# Find the province with the highest and lowest number of bridges in different conditions
province_statistics['Total'] = province_statistics.sum(axis=1)
highest_province = province_statistics['Total'].idxmax()
lowest_province = province_statistics['Total'].idxmin()

print("Statistics for Each Province:")
print(province_statistics)

print("\nProvince with the Highest Number of Bridges (Total):", highest_province)
print("Province with the Lowest Number of Bridges (Total):", lowest_province)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by province and calculate the percentage of bridges in each condition category
province_condition_percentages = province_statistics.div(province_statistics['Total'], axis=0) * 100

# Plot bar charts for each condition category
condition_categories = ['Jml_Baik', 'Jml_RR', 'Jml_RB', 'Jml_Kritis', 'Jml_Runtuh/Putus']
for category in condition_categories:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=province_condition_percentages.index, y=province_condition_percentages[category])
    plt.xticks(rotation=90)
    plt.title(f'Persentase Jembatan dalam Kategori Kondisi "{category}" per Provinsi')
    plt.ylabel('Persentase')
    plt.show()
```
